```python
import pandas as pd
```

```python
url="datasets/spotify_data.csv"
data=pd.read_csv(url)
data.head()
```

```python
hits=[]
flops=[]
x=[]

hits = data.loc[data['target'] == 1.0, ['track']]
flops = data.loc[data['target'] == 0.0, ['track']]
```

```python
hitwords=[]
flopwords=[]

for word in hits['track']:
    x = word.split()
    for w in x:
        hitwords.append(w)
hitwords
```

```python
for word in flops['track']:
    x = word.split()
    for w in x:
        flopwords.append(w)
flopwords
```

```python
hitlist = pd.Series(hitwords)

tophitlist=hitlist.value_counts()
tophitlist.head(20)
```

```python
floplist = pd.Series(flopwords)

topfloplist=floplist.value_counts()
topfloplist.head(20)
```

```python
topfloplist_df=topfloplist.to_frame()
tophitlist_df=tophitlist.to_frame()
```

```python
tophitlist_df.reset_index(level=0, inplace=True)
topfloplist_df.reset_index(level=0, inplace=True)
```

```python
tophit_df=tophitlist_df.rename(columns={'index': 'Hit Word', 0: 'Hit Count'})
topflop_df=topfloplist_df.rename(columns={'index': 'Flop Word', 0: 'Flop Count'})
```

```python
tophit_df
```

```python
merged_data = tophit_df.merge(topflop_df, left_index=True, right_index=True)
merged_data.head(20)
```

```python
merged_data_20=merged_data.iloc[:20]
merged_data_20
```

```python
merged_data_20.to_csv('hit_flop_words.csv')
```
