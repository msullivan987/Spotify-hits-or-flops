```python
tracks = results["tracks"]["items"]

all_new_tracks = []
all_new_artists = []
all_new_uris = []

for song in tracks:
    try:
        all_new_tracks.append(song["track"]["name"])
        all_new_artists.append(song["track"]["album"]["artists"][0]["name"])
        all_new_uris.append(song["track"]["uri"])
    except TypeError:
        pass
```

```python
new_songs_dict = {
    "Artist":all_new_artists,
    "Tracks":all_new_tracks,
    "URI":all_new_uris
}
```

```python
import pandas as pd

new_songs_df = pd.DataFrame(new_songs_dict)
new_songs_df.head()
```

```python
# Testing to see if this is how you pull the features we're looking for - hint: it is!

results = sp.audio_features("spotify:track:472vIK1ldetTxRxG3ovaiY")
```

```python
audio_features_list = []
for uri in all_new_uris:
    audio_features_list.append(sp.audio_features(uri))
```

```python
audio_features = [dictionary[0] for dictionary in audio_features_list]
```

```python
features_df = pd.DataFrame(audio_features)
features_df.head()
```

```python
features = features_df.drop(['uri', 'type', 'id', 'track_href', 'analysis_url'], axis=1)
features
```

```python
from tensorflow.keras.models import load_model
model = load_model("model_trained_2010s_13_feat.h5")
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(features)
X = X_scaler.transform(features)
```

```python
new_music_hit_prediction = model.predict_classes(X)
new_music_hit_prediction
```

```python
new_songs_df["Hit or Flop"]=new_music_hit_prediction
```

```python
new_song_hits_df = new_songs_df.drop('URI', axis=1)
new_song_hits_df = new_song_hits_df.replace({
    0:"Flop",
    1:"Hit"
})
new_song_hits_df.head()
```

```python
only_hits_df = new_song_hits_df.loc[new_song_hits_df["Hit or Flop"] == "Hit", :]
only_hits_df
```
